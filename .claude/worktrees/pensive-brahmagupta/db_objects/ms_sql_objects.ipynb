# [dbo].[FN05_CalcMonthlyPayment]

```sql
/****** Object:  UserDefinedFunction [dbo].[FN05_CalcMonthlyPayment]    Script Date: 2026-04-07 오후 2:51:09 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE FUNCTION [dbo].[FN05_CalcMonthlyPayment]
(
    @principal   DECIMAL(18,0),
    @annual_rate DECIMAL(6,4),
    @term_month  SMALLINT
)
RETURNS DECIMAL(18,2)
AS
BEGIN
    DECLARE @r DECIMAL(18,10) = @annual_rate / 100.0 / 12.0;
    IF @r = 0 RETURN ROUND(@principal / @term_month, 2);
    RETURN ROUND(
        @principal * @r * POWER(CAST(1+@r AS FLOAT), @term_month)
        / (POWER(CAST(1+@r AS FLOAT), @term_month) - 1), 2);
END;

GO
```

# [dbo].[FN07_AddBusinessDays]

```sql
/****** Object:  UserDefinedFunction [dbo].[FN07_AddBusinessDays]    Script Date: 2026-04-07 오후 2:51:09 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE FUNCTION [dbo].[FN07_AddBusinessDays]
(
    @start_dt   DATE,
    @add_days   INT
)
RETURNS DATE
AS
BEGIN
    DECLARE @result DATE = @start_dt;
    DECLARE @cnt    INT  = 0;
    WHILE @cnt < ABS(@add_days)
    BEGIN
        SET @result = DATEADD(DAY, SIGN(@add_days), @result);
        --  
        IF DATEPART(WEEKDAY, @result) <> 1 AND DATEPART(WEEKDAY, @result) <> 7  -- 1:, 7:
            SET @cnt = @cnt + 1;
    END;
    RETURN @result;
END;

GO
```

# [dbo].[FN09_FormatPhoneNo]

```sql
/****** Object:  UserDefinedFunction [dbo].[FN09_FormatPhoneNo]    Script Date: 2026-04-07 오후 2:51:09 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE FUNCTION [dbo].[FN09_FormatPhoneNo](@phone_no VARCHAR(20))
RETURNS VARCHAR(20)
AS
BEGIN
    IF @phone_no IS NULL RETURN NULL;
    DECLARE @digits VARCHAR(20) = '';
    DECLARE @i INT = 1;
    WHILE @i <= LEN(@phone_no)
    BEGIN
        IF SUBSTRING(@phone_no, @i, 1) BETWEEN '0' AND '9'
            SET @digits = @digits + SUBSTRING(@phone_no, @i, 1);
        SET @i = @i + 1;
    END;
    RETURN CASE LEN(@digits)
        WHEN 10 THEN LEFT(@digits,3) + '-' + SUBSTRING(@digits,4,3) + '-' + RIGHT(@digits,4)
        WHEN 11 THEN LEFT(@digits,3) + '-' + SUBSTRING(@digits,4,4) + '-' + RIGHT(@digits,4)
        ELSE @phone_no
    END;
END;

GO
```

# [dbo].[V02_OverdueStatus]

```sql
/****** Object:  View [dbo].[V02_OverdueStatus]    Script Date: 2026-04-07 오후 2:51:09 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE VIEW [dbo].[V02_OverdueStatus]
AS
SELECT
    s.loan_id,
    l.loan_no,
    c.cust_no,
    c.cust_nm,
    c.mobile_no,
    p.prod_nm,
    s.seq_no,
    s.due_dt,
    s.total_amt,
    s.paid_amt,
    ISNULL(s.total_amt, 0) - ISNULL(s.paid_amt, 0) AS unpaid_amt,
    s.overdue_days,
    s.overdue_int,
    CASE
        WHEN s.overdue_days BETWEEN 1  AND 30  THEN 'N1'
        WHEN s.overdue_days BETWEEN 31 AND 90  THEN 'N2'
        WHEN s.overdue_days BETWEEN 91 AND 180 THEN 'N3'
        WHEN s.overdue_days > 180              THEN 'N4'
        ELSE 'OK'
    END AS overdue_class
FROM [dbo].[E02_RepaySchedule] s
JOIN [dbo].[E01_LoanContract]  l ON s.loan_id = l.loan_id
JOIN [dbo].[B01_Customer]      c ON l.cust_id  = c.cust_id
JOIN [dbo].[C01_Product]       p ON l.prod_id  = p.prod_id
WHERE s.status_cd = 'OV';

GO
```

# [dbo].[V04_MonthlyStats]

```sql
/****** Object:  View [dbo].[V04_MonthlyStats]    Script Date: 2026-04-07 오후 2:51:09 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE VIEW [dbo].[V04_MonthlyStats]
AS
WITH monthly_execute AS (
    SELECT
        CONVERT(VARCHAR(7), execute_dt, 120) AS base_ym,
        prod_id,
        COUNT(*)                        AS execute_cnt,
        SUM(principal_amt)              AS execute_amt,
        AVG(final_rate)                 AS avg_rate
    FROM [dbo].[E01_LoanContract]
    WHERE status_cd IN ('ACT','CLS') AND execute_dt IS NOT NULL
    GROUP BY CONVERT(VARCHAR(7), execute_dt, 120), prod_id
),
monthly_repay AS (
    SELECT
        CONVERT(VARCHAR(7), tx_dt, 120)  AS base_ym,
        SUM(principal_amt)              AS repay_principal,
        SUM(interest_amt)               AS interest_income,
        COUNT(*)                        AS tx_cnt
    FROM [dbo].[F01_Transaction]
    WHERE tx_tp = 'PAY' AND result_cd = 'OK'
    GROUP BY CONVERT(VARCHAR(7), tx_dt, 120)
)
SELECT
    e.base_ym,
    e.prod_id,
    e.execute_cnt,
    e.execute_amt,
    e.avg_rate,
    ISNULL(r.repay_principal, 0) AS repay_principal,
    ISNULL(r.interest_income, 0) AS interest_income,
    ISNULL(r.tx_cnt, 0)          AS repay_tx_cnt
FROM monthly_execute e
LEFT JOIN monthly_repay r ON e.base_ym = r.base_ym;

GO
```

# [dbo].[V09_VipCustomer]

```sql
/****** Object:  View [dbo].[V09_VipCustomer]    Script Date: 2026-04-07 오후 2:51:09 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE VIEW [dbo].[V09_VipCustomer]
AS
SELECT
    c.cust_id,
    c.cust_no,
    c.cust_nm,
    c.credit_grade,
    c.joined_dt,
    DATEDIFF(YEAR, c.joined_dt, GETDATE()) AS tenure_years,
    cnt.active_loans,
    cnt.total_balance,
    cnt.total_interest_paid,
    seg.seg_cd,
    seg.rfm_total
FROM [dbo].[B01_Customer] c
CROSS APPLY (
    SELECT
        COUNT(DISTINCT l.loan_id) AS active_loans,
        SUM(l.principal_amt)      AS total_balance,
        SUM(t.interest_amt)       AS total_interest_paid
    FROM [dbo].[E01_LoanContract]  l
    LEFT JOIN [dbo].[F01_Transaction] t ON l.loan_id = t.loan_id AND t.tx_tp = 'PAY'
    WHERE l.cust_id = c.cust_id AND l.status_cd = 'ACT'
) cnt
LEFT JOIN (
    SELECT cust_id, seg_cd, rfm_total,
           ROW_NUMBER() OVER (PARTITION BY cust_id ORDER BY created_dt DESC) AS rn
    FROM [dbo].[I03_CustomerSegment]
) seg ON c.cust_id = seg.cust_id AND seg.rn = 1
WHERE c.vip_yn = 1;

GO
```

# [dbo].[TVF02_LoanTransactions]

```sql
/****** Object:  UserDefinedFunction [dbo].[TVF02_LoanTransactions]    Script Date: 2026-04-07 오후 2:51:09 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE FUNCTION [dbo].[TVF02_LoanTransactions]
(
    @loan_id    BIGINT,
    @fr_dt      DATE,
    @to_dt      DATE
)
RETURNS TABLE
AS
RETURN
(
    SELECT
        t.[tx_id],
        t.[tx_no],
        t.[tx_tp],
        t.[tx_dt],
        t.[tx_amt],
        t.[principal_amt],
        t.[interest_amt],
        t.[fee_amt],
        t.[pay_method],
        t.[result_cd],
        SUM(t.[principal_amt]) OVER (ORDER BY t.[tx_dt] ROWS UNBOUNDED PRECEDING) AS cumulative_principal,
        SUM(t.[interest_amt])  OVER (ORDER BY t.[tx_dt] ROWS UNBOUNDED PRECEDING) AS cumulative_interest
    FROM [dbo].[F01_Transaction] t
    WHERE t.[loan_id]  = @loan_id
      AND t.[tx_dt]   >= @fr_dt
      AND t.[tx_dt]   <  DATEADD(DAY, 1, @to_dt)
      AND t.[result_cd] = 'OK'
);

GO
```

# [dbo].[SP01_UpsertCustomer]

```sql
/****** Object:  StoredProcedure [dbo].[SP01_UpsertCustomer]    Script Date: 2026-04-07 오후 2:51:09 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[SP01_UpsertCustomer]
    @cust_no        CHAR(15),
    @cust_tp        CHAR(1)       = 'I',
    @cust_nm        NVARCHAR(100),
    @birth_dt       DATE          = NULL,
    @gender_cd      CHAR(1)       = NULL,
    @email          VARCHAR(255)  = NULL,
    @mobile_no      VARCHAR(20)   = NULL,
    @region_cd      CHAR(6)       = NULL,
    @operator       VARCHAR(50)   = NULL,
    @cust_id        BIGINT        = NULL OUTPUT,
    @result_msg     NVARCHAR(200) = NULL OUTPUT
AS
BEGIN
    SET NOCOUNT ON;
    SET XACT_ABORT ON;

    DECLARE @err_msg NVARCHAR(200);

    BEGIN TRY
        BEGIN TRANSACTION;

        IF EXISTS (SELECT 1 FROM [dbo].[B01_Customer] WHERE [cust_no] = @cust_no)
        BEGIN
            -- 
            UPDATE [dbo].[B01_Customer] SET
                [cust_nm]    = @cust_nm,
                [birth_dt]   = ISNULL(@birth_dt, [birth_dt]),
                [gender_cd]  = ISNULL(@gender_cd, [gender_cd]),
                [email]      = ISNULL(@email, [email]),
                [mobile_no]  = ISNULL(@mobile_no, [mobile_no]),
                [region_cd]  = ISNULL(@region_cd, [region_cd]),
                [updated_dt] = SYSDATETIME(),
                [updated_by] = ISNULL(@operator, SYSTEM_USER)
            WHERE [cust_no] = @cust_no;

            SELECT @cust_id = [cust_id] FROM [dbo].[B01_Customer] WHERE [cust_no] = @cust_no;
            SET @result_msg = 'Customer updated';
        END
        ELSE
        BEGIN
            --  
            INSERT INTO [dbo].[B01_Customer]
                ([cust_no],[cust_tp],[cust_nm],[birth_dt],[gender_cd],[email],[mobile_no],[region_cd],[created_by])
            VALUES
                (@cust_no, @cust_tp, @cust_nm, @birth_dt, @gender_cd, @email, @mobile_no, @region_cd,
                 ISNULL(@operator, SYSTEM_USER));

            SET @cust_id = SCOPE_IDENTITY();
            SET @result_msg = 'New customer: ' + CAST(@cust_id AS NVARCHAR) + N')';
        END

        COMMIT TRANSACTION;
    END TRY
    BEGIN CATCH
        IF @@TRANCOUNT > 0 ROLLBACK TRANSACTION;
        SET @result_msg = 'Error: ' + ERROR_MESSAGE();
        THROW;
    END CATCH
END;

GO
```

# [dbo].[SP02_SubmitLoanApplication]

```sql
/****** Object:  StoredProcedure [dbo].[SP02_SubmitLoanApplication]    Script Date: 2026-04-07 오후 2:51:09 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[SP02_SubmitLoanApplication]
    @cust_id      BIGINT,
    @prod_id      INT,
    @appl_amt     DECIMAL(18,0),
    @appl_term    SMALLINT,
    @purpose_cd   CHAR(4)       = NULL,
    @appl_channel CHAR(3)       = 'WEB',
    @emp_id       VARCHAR(50)   = NULL,
    @appl_id      BIGINT        = NULL OUTPUT,
    @appl_no      VARCHAR(20)   = NULL OUTPUT
AS
BEGIN
    SET NOCOUNT ON;
    SET XACT_ABORT ON;

    DECLARE @min_amt DECIMAL(18,0), @max_amt DECIMAL(18,0);
    DECLARE @min_term SMALLINT, @max_term SMALLINT;

    BEGIN TRY
        BEGIN TRANSACTION;

        --   
        IF NOT EXISTS (SELECT 1 FROM [dbo].[B01_Customer] WHERE [cust_id] = @cust_id AND [status_cd] = 'AC')
            THROW 50001, 'Invalid customer.', 1;

        --  
        IF EXISTS (SELECT 1 FROM [dbo].[B01_Customer] WHERE [cust_id] = @cust_id AND [blacklist_yn] = 1)
            THROW 50002, 'Customer restricted.', 1;

        --   
        SELECT @min_amt = [min_amt], @max_amt = [max_amt], @min_term = [min_term_month], @max_term = [max_term_month]
        FROM [dbo].[C01_Product]
        WHERE [prod_id] = @prod_id AND [status_cd] = 'AC';

        IF @@ROWCOUNT = 0 THROW 50003, 'Invalid product.', 1;
        IF @appl_amt < @min_amt OR @appl_amt > @max_amt THROW 50004, 'Amount over limit.', 1;
        IF @appl_term < @min_term OR @appl_term > @max_term THROW 50005, 'Term out of range.', 1;

        --  
        SET @appl_no = 'APL' + FORMAT(GETDATE(), 'yyyyMMdd') + RIGHT('000000' + CAST(@@SPID * 1000 + ABS(CHECKSUM(NEWID())) % 1000 AS VARCHAR), 6);

        --  
        INSERT INTO [dbo].[D01_LoanApplication]
            ([appl_no],[cust_id],[prod_id],[appl_amt],[appl_term],[purpose_cd],[appl_channel],[status_cd],[created_by])
        VALUES
            (@appl_no, @cust_id, @prod_id, @appl_amt, @appl_term, @purpose_cd, @appl_channel, 'REQ',
             ISNULL(@emp_id, SYSTEM_USER));

        SET @appl_id = SCOPE_IDENTITY();

        --    (  )
        INSERT INTO [dbo].[L01_CreditInquiryLog] ([cust_id],[appl_id],[inquiry_tp],[result_cd])
        VALUES (@cust_id, @appl_id, 'CB', 'OK');

        COMMIT TRANSACTION;
    END TRY
    BEGIN CATCH
        IF @@TRANCOUNT > 0 ROLLBACK TRANSACTION;
        THROW;
    END CATCH
END;

GO
```

# [dbo].[SP03_GenerateRepaySchedule]

```sql
/****** Object:  StoredProcedure [dbo].[SP03_GenerateRepaySchedule]    Script Date: 2026-04-07 오후 2:51:09 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[SP03_GenerateRepaySchedule]
    @loan_id    BIGINT,
    @result_cnt INT = NULL OUTPUT
AS
BEGIN
    SET NOCOUNT ON;
    SET XACT_ABORT ON;

    DECLARE
        @principal    DECIMAL(18,0),
        @rate         DECIMAL(6,4),
        @term         SMALLINT,
        @repay_tp     CHAR(2),
        @repay_day    TINYINT,
        @execute_dt   DATE,
        @monthly_rate DECIMAL(18,10),
        @monthly_pay  DECIMAL(18,2),
        @balance      DECIMAL(18,2),
        @seq          SMALLINT,
        @due_dt       DATE,
        @principal_pay DECIMAL(18,2),
        @interest_pay  DECIMAL(18,2);

    --   
    SELECT
        @principal = [principal_amt], @rate = [final_rate], @term = [term_month],
        @repay_tp = [repay_tp], @repay_day = [repay_day], @execute_dt = [execute_dt]
    FROM [dbo].[E01_LoanContract]
    WHERE [loan_id] = @loan_id;

    IF @@ROWCOUNT = 0 THROW 50010, 'Loan not found.', 1;

    --   
    DELETE FROM [dbo].[E02_RepaySchedule] WHERE [loan_id] = @loan_id;

    SET @monthly_rate = @rate / 100 / 12;
    SET @balance      = @principal;

    --    
    IF @repay_tp = 'EM' AND @monthly_rate > 0
        SET @monthly_pay = @principal * @monthly_rate * POWER(1 + @monthly_rate, @term)
                           / (POWER(1 + @monthly_rate, @term) - 1);
    ELSE
        SET @monthly_pay = @principal / @term;

    BEGIN TRY
        BEGIN TRANSACTION;

        SET @seq = 1;
        WHILE @seq <= @term
        BEGIN
            SET @due_dt = DATEADD(DAY, @repay_day - DAY(@execute_dt),
                            DATEADD(MONTH, @seq, @execute_dt));

            IF @repay_tp = 'EM'
            BEGIN
                SET @interest_pay  = ROUND(@balance * @monthly_rate, 2);
                SET @principal_pay = ROUND(CASE WHEN @seq = @term THEN @balance
                                           ELSE @monthly_pay - @interest_pay END, 2);
            END
            ELSE IF @repay_tp = 'PM'
            BEGIN
                SET @principal_pay = ROUND(@principal / @term, 2);
                SET @interest_pay  = ROUND(@balance * @monthly_rate, 2);
            END
            ELSE -- BL 
            BEGIN
                SET @principal_pay = CASE WHEN @seq = @term THEN @balance ELSE 0 END;
                SET @interest_pay  = ROUND(@principal * @monthly_rate, 2);
            END;

            SET @balance = @balance - @principal_pay;
            IF @balance < 0 SET @balance = 0;

            INSERT INTO [dbo].[E02_RepaySchedule]
                ([loan_id],[seq_no],[due_dt],[principal_amt],[interest_amt],[balance_amt])
            VALUES
                (@loan_id, @seq, @due_dt, @principal_pay, @interest_pay, @balance);

            SET @seq = @seq + 1;
        END;

        SET @result_cnt = @term;
        COMMIT TRANSACTION;
    END TRY
    BEGIN CATCH
        IF @@TRANCOUNT > 0 ROLLBACK TRANSACTION;
        THROW;
    END CATCH
END;

GO
```

# [dbo].[SP04_ProcessOverdue]

```sql
/****** Object:  StoredProcedure [dbo].[SP04_ProcessOverdue]    Script Date: 2026-04-07 오후 2:51:09 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[SP04_ProcessOverdue]
    @base_dt    DATE = NULL,
    @result_cnt INT  = NULL OUTPUT
AS
BEGIN
    SET NOCOUNT ON;
    IF @base_dt IS NULL SET @base_dt = CAST(GETDATE() AS DATE);

    BEGIN TRY
        BEGIN TRANSACTION;

        --   
        UPDATE s SET
            s.[status_cd]    = 'OV',
            s.[overdue_days] = DATEDIFF(DAY, s.[due_dt], @base_dt),
            s.[overdue_int]  = ROUND(
                (ISNULL(s.[total_amt], 0) - ISNULL(s.[paid_amt], 0))
                * (l.[final_rate] / 100 / 365 * DATEDIFF(DAY, s.[due_dt], @base_dt)), 2)
        FROM [dbo].[E02_RepaySchedule] s
        JOIN [dbo].[E01_LoanContract]  l ON s.[loan_id] = l.[loan_id]
        WHERE s.[due_dt]     < @base_dt
          AND s.[status_cd]  = 'SC'
          AND l.[status_cd]  = 'ACT';

        SET @result_cnt = @@ROWCOUNT;

        --    
        INSERT INTO [dbo].[F02_OverdueManagement] ([loan_id],[overdue_start_dt],[overdue_amt],[status_cd])
        SELECT DISTINCT s.[loan_id], MIN(s.[due_dt]), SUM(s.[total_amt] - ISNULL(s.[paid_amt],0)), 'AC'
        FROM [dbo].[E02_RepaySchedule] s
        WHERE s.[status_cd] = 'OV'
          AND NOT EXISTS (
              SELECT 1 FROM [dbo].[F02_OverdueManagement] o
              WHERE o.[loan_id] = s.[loan_id] AND o.[status_cd] = 'AC')
        GROUP BY s.[loan_id];

        COMMIT TRANSACTION;
    END TRY
    BEGIN CATCH
        IF @@TRANCOUNT > 0 ROLLBACK TRANSACTION;
        THROW;
    END CATCH
END;

GO
```

# [dbo].[SP05_ProcessPayment]

```sql
/****** Object:  StoredProcedure [dbo].[SP05_ProcessPayment]    Script Date: 2026-04-07 오후 2:51:09 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[SP05_ProcessPayment]
    @loan_id     BIGINT,
    @pay_amt     DECIMAL(18,2),
    @pay_method  CHAR(3)       = 'ARS',
    @bank_cd     CHAR(4)       = NULL,
    @account_no  VARCHAR(30)   = NULL,
    @batch_id    VARCHAR(50)   = NULL,
    @tx_id       BIGINT        = NULL OUTPUT
AS
BEGIN
    SET NOCOUNT ON;
    SET XACT_ABORT ON;

    DECLARE @tx_no       VARCHAR(30),
            @principal   DECIMAL(18,2),
            @interest    DECIMAL(18,2),
            @remain      DECIMAL(18,2),
            @sched_id    BIGINT;

    BEGIN TRY
        BEGIN TRANSACTION;

        --     
        SELECT TOP 1
            @sched_id  = [sched_id],
            @principal = [principal_amt],
            @interest  = [interest_amt]
        FROM [dbo].[E02_RepaySchedule]
        WHERE [loan_id] = @loan_id AND [status_cd] IN ('SC','OV')
        ORDER BY [due_dt] ASC;

        IF @@ROWCOUNT = 0 THROW 50020, 'No schedule to pay.', 1;

        SET @remain = @pay_amt;

        --   
        SET @interest  = CASE WHEN @remain >= @interest  THEN @interest  ELSE @remain END;
        SET @remain    = @remain - @interest;
        SET @principal = CASE WHEN @remain >= @principal THEN @principal ELSE @remain END;

        --   
        SET @tx_no = 'TX' + FORMAT(SYSDATETIME(), 'yyyyMMddHHmmssfff');

        --  
        INSERT INTO [dbo].[F01_Transaction]
            ([tx_no],[loan_id],[tx_tp],[tx_amt],[principal_amt],[interest_amt],
             [pay_method],[bank_cd],[account_no],[batch_id],[result_cd])
        VALUES
            (@tx_no, @loan_id, 'PAY', @pay_amt, @principal, @interest,
             @pay_method, @bank_cd, @account_no, @batch_id, 'OK');

        SET @tx_id = SCOPE_IDENTITY();

        --  
        UPDATE [dbo].[E02_RepaySchedule] SET
            [status_cd] = 'PD',
            [paid_dt]   = CAST(GETDATE() AS DATE),
            [paid_amt]  = @pay_amt,
            [overdue_days] = 0
        WHERE [sched_id] = @sched_id;

        COMMIT TRANSACTION;
    END TRY
    BEGIN CATCH
        IF @@TRANCOUNT > 0 ROLLBACK TRANSACTION;
        THROW;
    END CATCH
END;

GO
```

# [dbo].[SP06_AggDailyPerformance]

```sql
/****** Object:  StoredProcedure [dbo].[SP06_AggDailyPerformance]    Script Date: 2026-04-07 오후 2:51:09 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[SP06_AggDailyPerformance]
    @base_dt DATE = NULL,
    @result_cnt INT = NULL OUTPUT
AS
BEGIN
    SET NOCOUNT ON;
    IF @base_dt IS NULL SET @base_dt = CAST(GETDATE() AS DATE);

    --   
    DELETE FROM [dbo].[K01_DailyPerformance] WHERE [base_dt] = @base_dt;

    --  INSERT
    INSERT INTO [dbo].[K01_DailyPerformance]
        ([base_dt],[branch_cd],[prod_tp],
         [new_appl_cnt],[new_approve_cnt],[new_execute_cnt],[new_execute_amt],
         [total_balance],[total_loan_cnt],[overdue_cnt],[overdue_amt],[interest_income])
    SELECT
        @base_dt,
        ISNULL(l.[branch_cd], 'HQ'),
        p.[prod_tp],
        COUNT(DISTINCT a.[appl_id]),
        COUNT(DISTINCT CASE WHEN a.[status_cd] IN ('AP','EX') THEN a.[appl_id] END),
        COUNT(DISTINCT CASE WHEN l.[execute_dt] = @base_dt THEN l.[loan_id] END),
        ISNULL(SUM(CASE WHEN l.[execute_dt] = @base_dt THEN l.[principal_amt] ELSE 0 END), 0),
        ISNULL(SUM(CASE WHEN l.[status_cd] = 'ACT' THEN l.[principal_amt] ELSE 0 END), 0),
        COUNT(DISTINCT CASE WHEN l.[status_cd] = 'ACT' THEN l.[loan_id] END),
        COUNT(DISTINCT CASE WHEN s.[status_cd] = 'OV' THEN s.[loan_id] END),
        ISNULL(SUM(CASE WHEN s.[status_cd] = 'OV' THEN s.[total_amt] - ISNULL(s.[paid_amt],0) ELSE 0 END), 0),
        ISNULL(SUM(CASE WHEN t.[tx_dt] >= @base_dt AND t.[tx_dt] < DATEADD(DAY,1,@base_dt) THEN t.[interest_amt] ELSE 0 END), 0)
    FROM [dbo].[C01_Product]              p
    LEFT JOIN [dbo].[D01_LoanApplication]  a ON p.[prod_id] = a.[prod_id]
    LEFT JOIN [dbo].[E01_LoanContract]     l ON p.[prod_id] = l.[prod_id]
    LEFT JOIN [dbo].[E02_RepaySchedule]    s ON l.[loan_id] = s.[loan_id]
    LEFT JOIN [dbo].[F01_Transaction]      t ON l.[loan_id] = t.[loan_id]
    GROUP BY ISNULL(l.[branch_cd],'HQ'), p.[prod_tp];

    SET @result_cnt = @@ROWCOUNT;
END;

GO
```

# [dbo].[SP08_CalcLoanLossProvision]

```sql
/****** Object:  StoredProcedure [dbo].[SP08_CalcLoanLossProvision]    Script Date: 2026-04-07 오후 2:51:09 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[SP08_CalcLoanLossProvision]
    @base_ym    CHAR(6),
    @result_cnt INT = NULL OUTPUT
AS
BEGIN
    SET NOCOUNT ON;
    SET XACT_ABORT ON;

    DECLARE
        @loan_id      BIGINT,
        @class_cd     CHAR(2),
        @balance      DECIMAL(18,0),
        @prov_rate    DECIMAL(8,6),
        @prov_amt     DECIMAL(18,0),
        @cnt          INT = 0;

    --   
    DELETE FROM [dbo].[G02_LoanLossProvision] WHERE [base_ym] = @base_ym;

    BEGIN TRY
        BEGIN TRANSACTION;

        --  
        DECLARE cur_asset CURSOR FAST_FORWARD FOR
            SELECT
                l.[loan_id],
                ISNULL(g.[class_cd], 'N'),
                l.[principal_amt]
            FROM [dbo].[E01_LoanContract] l
            LEFT JOIN (
                SELECT [loan_id], [class_cd]
                FROM [dbo].[G01_AssetClassification]
                WHERE [class_ym] = @base_ym
            ) g ON l.[loan_id] = g.[loan_id]
            WHERE l.[status_cd] = 'ACT';

        OPEN cur_asset;
        FETCH NEXT FROM cur_asset INTO @loan_id, @class_cd, @balance;

        WHILE @@FETCH_STATUS = 0
        BEGIN
            --  
            SET @prov_rate = CASE @class_cd
                WHEN 'N'  THEN 0.005
                WHEN 'L'  THEN 0.02
                WHEN 'D3' THEN 0.20
                WHEN 'D4' THEN 0.50
                WHEN 'D5' THEN 1.00
                ELSE 0.005
            END;

            SET @prov_amt = CAST(@balance * @prov_rate AS DECIMAL(18,0));

            INSERT INTO [dbo].[G02_LoanLossProvision]
                ([base_ym],[loan_id],[class_cd],[balance_amt],[provision_rate],[provision_amt])
            VALUES
                (@base_ym, @loan_id, @class_cd, @balance, @prov_rate, @prov_amt);

            SET @cnt = @cnt + 1;
            FETCH NEXT FROM cur_asset INTO @loan_id, @class_cd, @balance;
        END;

        CLOSE cur_asset;
        DEALLOCATE cur_asset;

        SET @result_cnt = @cnt;
        COMMIT TRANSACTION;
    END TRY
    BEGIN CATCH
        IF CURSOR_STATUS('local', 'cur_asset') >= 0
        BEGIN
            CLOSE cur_asset;
            DEALLOCATE cur_asset;
        END
        IF @@TRANCOUNT > 0 ROLLBACK TRANSACTION;
        THROW;
    END CATCH
END;

GO
```

```sql
/****** Object:  StoredProcedure [dbo].[SP09_BatchJobControl]    Script Date: 2026-04-07 오후 2:51:09 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[SP09_BatchJobControl]
    @job_cd      VARCHAR(30),
    @action      CHAR(2),        -- ST:, EN:, ER:
    @processed   INT          = 0,
    @success     INT          = 0,
    @fail        INT          = 0,
    @error_msg   NVARCHAR(MAX)= NULL,
    @param_json  NVARCHAR(MAX)= NULL,
    @run_id      BIGINT       = NULL OUTPUT
AS
BEGIN
    SET NOCOUNT ON;

    IF @action = 'ST'
    BEGIN
        INSERT INTO [dbo].[M02_BatchRunHistory]
            ([job_cd],[run_dt],[status_cd],[run_by],[param_json])
        VALUES
            (@job_cd, SYSDATETIME(), 'RN', SYSTEM_USER, @param_json);
        SET @run_id = SCOPE_IDENTITY();
    END
    ELSE IF @action IN ('EN','ER')
    BEGIN
        UPDATE [dbo].[M02_BatchRunHistory] SET
            [end_dt]        = SYSDATETIME(),
            [status_cd]     = CASE @action WHEN 'EN' THEN 'OK' ELSE 'ER' END,
            [processed_cnt] = @processed,
            [success_cnt]   = @success,
            [fail_cnt]      = @fail,
            [error_msg]     = @error_msg
        WHERE [run_id] = @run_id;
    END
END;

GO
```
